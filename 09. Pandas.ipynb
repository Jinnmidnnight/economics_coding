## Series

A Series is a Pandas object that stores single column of **data**, with row labels for each observation. Pandas refers to the row labels as the **index** of the Series.

Let's create a Series which contains the share of population of age under 14 from 1970 to 2020 in 5 year interval. Actually we already used this data for the data visualization lecture.


```python
data = [42.5, 38.6, 34.0, 30.2, 25.6, 23.4, 21.1, 19.1, 16.1, 13.8, 12.2]
year = [1970, 1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015, 2020]

Share_Young = pd.Series(data, index=year, name="Share_Young")
```

## Data Frame

A DataFrame is a Pandas object that stores multiple columns of data. We can think a DataFrames a multiple Series stacked side by side as columns. It has a similar structure as an Excel spreadsheet. 

In addition to row labels (an index), DataFrames also have column labels which are refered as the column names.

Let's create a Dataframe which contains the shares of population according to three age groups: "Young" being of age under 14, "Middel" between 15 and 64, "Old" 65 and over. The data points are from 1970 to 2020 in 5 year interval.

## Working with DataFrame

### Read/write data from files

- In order to work analyze data, they should first be loaded into the computer memory.

- Usually data are stored in files on storage devices (hard disk, portable storages or even on the cloud) in various formats (fixed-width format texts, comma separated values, excel, jason, html, etc). 

- We need to load data from files into a Pandas dataframe. We also need to be able to save the process data in files for future use. As in the case of reading data, there are several possible formats in which Pandas save the data on files.

- For more info about input/output, refer to https://pandas.pydata.org/docs/reference/io.html

```python
# read data from an excel file

df = pd.read_excel("Data_Fig/EAP_All.xlsx")
df
```

```python
# read data from a csv file

df = pd.read_csv("Data_Fig/EAP_All.csv")
df
```

```python
# write data on file in html format

df.to_html("Data_Fig/EAP_All.html")
```

### Cleanup

```python
# remove redundant characters

df["Quarter"] = df["Quarter"].str.removesuffix("/4")
df
```

```python
# create new integer columns with the splited year and qtr

df[["Year", "Qtr"]] = df["Quarter"].str.split(".", expand=True).astype('int')
df
```

```python
# replace string and change data types

df["Pop"] = df["Pop"].str.replace(',', '').astype('int')
df["EAP"] = df["EAP"].str.replace(',', '').astype('int')
df["Empl"] = df["Empl"].str.replace(',', '').astype('int')

df.info()
```

### Filtering

```python
df["Year"] > 2010
```

```python
# filter dataframe using conditional statements

df[df["Year"] > 2010].head()
```

```python
df[(df["Year"] > 2010) & (df["Qtr"] == 1)].count()
```

```python
# unique values in a column

df["Age"].unique()
```

```python
# count observations for each unique value

df["Age"].value_counts()
```

```python
# change values of columns with filtering

df.loc[df["Age"] == "All", "Age"] = 0
df.loc[df["Age"] == "15_19", "Age"] = 1
df.loc[df["Age"] == "20_29", "Age"] = 2
df.loc[df["Age"] == "30_39", "Age"] = 3
df.loc[df["Age"] == "40_49", "Age"] = 4
df.loc[df["Age"] == "50_59", "Age"] = 5
df.loc[df["Age"] == "60_over", "Age"] = 6

# change data type to int

df["Age"] = df["Age"].astype('int')

df
```

## Analyzing data by group

We want to split the dataframe into several smaller sub-dataframe by characteristics of variables, then apply statistical analyses on each sub-dataframe. This can be conducted using `groupby` function in Pandas. For example, we want to compute the average unemployment rates for different age groups or the correlation coefficients of the unemployment rates and participation rates within each age group. For this task, we will first group the whole dataframe by `Age`.

```python
# compute the average unemployment rates for each age group without grouping

for i in range(7):
  print(df.loc[df["Age"] == i, "U_Rate"].mean())
```

```python
# compute the correlation coefficients for each age group without grouping

for i in range(7):
  print(df.loc[df["Age"] == i, ["Pop_growth", "P_Rate", "U_Rate"]].corr().round(2))
```

```python
# create group by `Age`

df_group = df.groupby("Age")

df_group.groups
```

```python
# size of each group

df_group.size()
```

```python
# number of observations of a specific variable in each group

df_group["Pop_growth"].count()
```

```python
# view data for a specific group

df_group.get_group(1)
```

```python
# equivalently

df[df["Age"]==1]
```

```python
df_group.describe()
```

```python
# averages of all variables for all groups

df_group.mean().round(2)
```

```python
# averages of all variables for a specific group

df_group.get_group(1).mean().round(2)
```

```python
# averages of specific variables for all groups

df_group[["Pop_growth", "P_Rate", "U_Rate"]].mean().round(2)
```

```python
# correlation coefficients of selected variables for each group

df_group[["Pop_growth", "P_Rate", "U_Rate"]].corr()
```

```python
# correlation coefficients of selected variables for a specific age group

df_group.get_group(1)[["Pop_growth", "P_Rate", "U_Rate"]].corr()
```

```python
# population growth rates

df.loc[df["Age"]==2, "Pop_growth"]
```

```python
# re-calculate the population growth rate by age group to correct mistakes at 2000.1

df["Pop_growth"] = df_group["Pop"].pct_change()*100
```

```python
plt.figure(figsize=(18, 5))

q = 1
plt.plot(df.loc[(df["Age"] == 1) & (df["Qtr"] == q), "Pop_growth"], label='15-19')
plt.plot(df.loc[(df["Age"] == 2) & (df["Qtr"] == q), "Pop_growth"], label='20-29')
plt.plot(df.loc[(df["Age"] == 3) & (df["Qtr"] == q), "Pop_growth"], label='30-39')
plt.plot(df.loc[(df["Age"] == 4) & (df["Qtr"] == q), "Pop_growth"], label='40-49')
plt.plot(df.loc[(df["Age"] == 5) & (df["Qtr"] == q), "Pop_growth"], label='50-59')
plt.plot(df.loc[(df["Age"] == 6) & (df["Qtr"] == q), "Pop_growth"], label='60 over')
plt.ylabel('%')
plt.title('Population Growth Rate')
plt.legend(loc='upper right')
plt.grid(True)
```

## Concatenate dataframes

pd.concat() is used to concatenate pandas objects such as DataFrames and Series into a single one. The following example displays how `concat()` works in various ways.

```python
# extract two small dataframes from df

df1 = df_group.get_group(1).loc["2000.1":"2001.4", ["Pop", "EAP", "Empl", "Year", "Qtr"]]
df2 = df_group.get_group(1).loc["2001.1":"2002.4", ["Year", "Qtr", "Pop_growth", "P_Rate", "U_Rate"]]
```

```python
df1
```

```python
df2
```

```python
# concatenate df1 and df2 in row-direction

pd.concat([df1, df2], axis=0, join="outer")
pd.concat([df1, df2], axis=0, ignore_index=True)
pd.concat([df1, df2], axis=0, join="inner")
```

```python
# concatenate df1 and df2 in column-direction

pd.concat([df1, df2], axis=1)
pd.concat([df1, df2], axis=1, join="inner")
```

For more info, refer to https://pandas.pydata.org/docs/reference/api/pandas.concat.html

## Merge dataframes

pd.merge() is used to merge two dataframes based on `key` on each dataframe. It is similar to pd.concat([df1, df2], axis=1), but provides more options to combine two dataframes. The following example displays how `merge()` works.

```python
# on=index or "Common columns"

pd.merge(df1, df2, how="inner", on="Quarter")
pd.merge(df1, df2, how="inner", left_index=True, right_index=True)
pd.merge(df1, df2, how="inner", on=["Year", "Qtr"])
```

```python
# how="outer"

pd.merge(df1, df2, how="outer", on="Quarter")
```

```python
# how="left"

pd.merge(df1, df2, how="left", left_on="Quarter", right_on="Quarter")
```

```python
# how="right"

pd.merge(df1, df2, how="right", on="Quarter")
pd.merge(df1, df2, how="right", on=["Year", "Qtr"])
```

There are other possible ways to merge multiple dataframes: outer, left, right, which will not be covered in this lecture. For more info about `merge`, refer to https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.merge.html
