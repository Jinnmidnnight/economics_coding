## Q1. Population Pyramid of Korea in 2021

Below is the population structure of Korea in 2021 by sex and age, so called the population pyramid. Investigate the following figures very carefully and write the Python codes to replicate all the details of the figures including legends, annotations and colors as closely as possible. For the list of named colors in Matplotlib library, refer to https://matplotlib.org/stable/gallery/color/named_colors.html

The data for this figure are provided below. The variables `male` and `female` contain the male and female population sizes in 21 age groups of 5 year interval. The data are taken from the KOSIS.

```python
# population data of Korea in 2021

age_group = ['0~4 years', '5~9 years', '10~14 years', '15~19 years', '20~24 years', '25~29 years', '30~34 years', '35~39 years', '40~44 years', '45~49 years', '50~54 years', '55~59 years', '60~64 years', '65~69 years', '70~74 years', '75~79 years', '80~84 years', '85~89 years', '90~94 years', '95~99 years', '100 and over']
male = [805852, 1133426, 1186319, 1198911, 1681183, 1954718, 1759768, 1877724, 2038116, 2100431, 2243979, 2064943, 2016321, 1433783, 975415, 685723, 447622, 189457, 47148, 8149, 1056]
female = [764557, 1078946, 1118371, 1114388, 1542471, 1708409, 1572504, 1744623, 1943594, 2033792, 2233654, 2032973, 2081537, 1542824, 1114229, 898881, 731214, 428224, 161610, 35719, 5507]
```

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# male, female 데이터 정리(반올림)
male_np = np.round(male, -3)
female_np = np.round(female, -3)

# female_left 리스트 생성
female_left = []
for f in female_np:
    female_left.append(-f)

# 인구 피라미드 그리기
fig = plt.figure(figsize=(12,8))
plt.barh(age_group, male, color="#3DE0D2", label="Male")
plt.barh(age_group, female_left, color="#EE82EF", label="Female")
for i in range(len(age_group)):
    plt.text(female_left[i], i, s="{:,.0f}  ".format(female_left[i]), ha="right", va="center")
    plt.text(male[i], i, s="  {:,.0f}".format(male[i]), ha="left", va="center")
for axis in ['top', 'bottom', 'left', 'right']:
    plt.gca().spines[axis].set_color('white')
plt.gca().get_xaxis().set_visible(False)
plt.tick_params(axis='y', pad=80, color='white')
plt.title('Population of Korea in 2021', fontsize=20, pad=10)
plt.legend(fontsize=12)
plt.show
```

## Q2. Income, Asset and Debt by Income Decile from the Household Survey

The following graphs show some summary statistics for income, asset holdings and debt by income Decile from the household survey (가계금융복지조사) in 2022. They display the averages of 
1. the tatal income, 
2. the share of disposable income and non-consumption income out of the total income, 
3. the total asset holdings
4. the share of financial and real asset out of the the total assets,
5. the age of the head of household
6. the total debt

Investigate the following figures very carefully and write the Python codes to replicate all the details of the figures including legends, annotations and colors as closely as possible. 


<img src="Income_Decile_2022.jpg" style="width:60%;">




The Data for these graphs are given below which are taken from KOSIS. The desciptions of variables are as follows.
- income_total: the average amount of total income for each income decile from 1 to 10. The unit is 10,000 KRW.
- income_disp: the average amount of disposable income for each income decile from 1 to 10. The unit is 10,000 KRW.
- income_no_cons: the average amount of non-consumption income (taxex, social insurance controbutions, etc.) for each income decile from 1 to 10. The unit is 10,000 KRW.
- age_head: the average age of the head of household. 
- asset_total: the average amount of total assets for each income decile from 1 to 10. The unit is 10,000 KRW.
- asset_financial: the average amount of financial assets for each income decile from 1 to 10. The unit is 10,000 KRW.
- asset_real: the average amount of real assets for each income decile from 1 to 10. The unit is 10,000 KRW.
- debt: the average amount of total debt for each income decile from 1 to 10. The unit is 10,000 KRW.


```python
# data from household survey

income_total = [897, 1749, 2624, 3537, 4491, 5582, 6863, 8436, 10909, 19042]
income_disp = [797, 1544, 2255, 3008, 3810, 4679, 5753, 6992, 8884, 14563]
income_no_cons = [101, 205, 370, 529, 681, 903, 1109, 1444, 2025, 4478]
age_head = [68.2, 63.1, 57.3, 54.6, 53.2, 51.5, 51.4, 50.1, 50.9, 51.5]
asset_total = [12407, 21970, 27511, 32508, 41889, 45692, 57223, 66604, 86386, 155475]
asset_real = [9484, 17562, 21487, 24528, 33018, 34527, 44277, 52475, 67519, 121542]
asset_financial = [2922, 4409, 6024, 7980, 8871, 11165, 12947, 14129, 18867, 33933]
debt = [1068, 2364, 3975, 5042, 6969, 8346, 9954, 12688, 15637, 25644]
```

```python
# your code here

# set figure and axes

fig, axes = plt.subplots(3, 2, figsize=(16, 24))
plt.title('Income, Asset and Debt by Income Decile', fontsize=20, pad=10)
income_decile = [i for i in range(1,11)]

# add charts on each axes

# Income
income_total_bar = []
for i in income_total:
    income_total_bar.append(float('%0.1f' % float(round(i,-1)/100)))
axes[0,0].bar(income_decile, income_total_bar, color="#FF7C52")
axes[0,0].set_xticks(income_decile)
axes[0,0].set_xlabel('Income Decile', fontsize=10)
axes[0,0].set_ylabel('million won', fontsize=10)
axes[0,0].set_title('Income', fontsize=15)
for axis in ['top','right']:
    axes[0,0].spines[axis].set_color('white')
for i in range(len(income_decile)):
    axes[0,0].text(income_decile[i], income_total_bar[i], income_total_bar[i],
                   fontsize = 10,
                   ha='center',
                   va='bottom',)

# Shares out of Total Income
disposable_bar = []
non_disposable_bar = []
for i in range(len(income_total)):
    disposable_bar.append(float('%0.1f' % float((income_disp[i]/income_total[i])*100)))
    non_disposable_bar.append(float('%0.1f' % float((income_no_cons[i]/income_total[i])*100)))
axes[0,1].bar(np.array(income_decile)-0.25, disposable_bar,width=0.4, color="#40DDCD", label='Disposable income')
axes[0,1].bar(np.array(income_decile)+0.25, non_disposable_bar,width=0.4, color="#FE66AE", label='Non-Disposable income')
axes[0,1].set_xticks([2,4,6,8,10])
axes[0,1].set_yticks([0,20,40,60,80,100])
axes[0,1].set_ylabel('million won', fontsize=10)
axes[0,1].set_title('Shares out of Total Income', fontsize=15)
axes[0,1].legend()
for axis in ['top','right']:
    axes[0,1].spines[axis].set_color('white')
for i in range(len(income_decile)):
    axes[0,1].text(income_decile[i], disposable_bar[i], disposable_bar[i],
                   fontsize = 10,
                   ha='right',
                   va='bottom',)
    axes[0,1].text(income_decile[i], non_disposable_bar[i], non_disposable_bar[i],
               fontsize = 10,
               ha='left',
               va='bottom',)

# Total Asset
total_asset_bar = []
for i in asset_total:
    total_asset_bar.append(float('%0.1f' % float(round(i,-1)/100)))
axes[1,0].bar(income_decile, total_asset_bar, color="#FEA500")
axes[1,0].set_xticks(income_decile)
axes[1,0].set_ylabel('million won', fontsize=10)
axes[1,0].set_title('Total Asset', fontsize=15)
for axis in ['top','right']:
    axes[1,0].spines[axis].set_color('white')
for i in range(len(income_decile)):
    axes[1,0].text(income_decile[i], total_asset_bar[i], total_asset_bar[i],
                   fontsize = 10,
                   ha='center',
                   va='bottom',)

# Shares out of Total Asset
real_bar = []
fin_bar = []
for i in range(len(asset_total)):
    real_bar.append(float('%0.1f' % float((asset_real[i]/asset_total[i])*100)))
    fin_bar.append(float('%0.1f' % float((asset_financial[i]/asset_total[i])*100)))
axes[1,1].bar(np.array(income_decile)-0.25, fin_bar,width=0.4, color="#81CFE5", label='Financial Asset')
axes[1,1].bar(np.array(income_decile)+0.25, real_bar,width=0.4, color="#FDD903", label='Real Asset')
axes[1,1].set_xticks([2,4,6,8,10])
axes[1,1].set_yticks([0,20,40,60,80,100])
axes[1,1].set_ylabel('million won', fontsize=10)
axes[1,1].set_title('Shares out of Total Asset', fontsize=15)
axes[1,1].legend()
for axis in ['top','right']:
    axes[1,1].spines[axis].set_color('white')
for i in range(len(income_decile)):
    axes[1,1].text(income_decile[i], real_bar[i], real_bar[i],
                   fontsize = 10,
                   ha='left',
                   va='bottom',)
    axes[1,1].text(income_decile[i], fin_bar[i], fin_bar[i],
               fontsize = 10,
               ha='right',
               va='bottom',)

# Age of Head of Household
age_head_bar = []
for i in age_head:
    age_head_bar.append(i)
axes[2,0].bar(income_decile, age_head_bar, color="#7D8000")
axes[2,0].set_xticks(income_decile)
axes[2,0].set_ylabel('year', fontsize=10)
axes[2,0].set_title('Age of Head of Household', fontsize=15)
for axis in ['top','right']:
    axes[2,0].spines[axis].set_color('white')
for i in range(len(income_decile)):
    axes[2,0].text(income_decile[i], age_head_bar[i], age_head_bar[i],
                   fontsize = 10,
                   ha='center',
                   va='bottom',)

# Debt
debt_bar = []
for i in debt:
    debt_bar.append(float('%0.1f' % float(round(i,-1)/100)))
axes[2,1].bar(income_decile, debt_bar, color="#01817F")
axes[2,1].set_xticks(income_decile)
axes[2,1].set_ylabel('million won', fontsize=10)
axes[2,1].set_title('Debt', fontsize=15)
for axis in ['top','right']:
    axes[2,1].spines[axis].set_color('white')
for i in range(len(income_decile)):
    axes[2,1].text(income_decile[i], debt_bar[i], debt_bar[i],
                   fontsize = 10,
                   ha='center',
                   va='bottom',)

plt.suptitle('Income, Asset and Debt by Income Decile, 2022', fontsize=20, va='bottom')
plt.tight_layout()
plt.show()
```
