## Q1. Write the Python codes to find the zeros of the following functions.

Q1-1. $$ F(x) = \int_{x^2}^\infty \frac{1}{\sqrt{2\pi}} e^{-t^2/2} dt + e^{x} - 3x^3 + 2x + 4$$

```python
# your code here

# F(x) 식 세우기
def integrand1_1(t):
    return np.exp(-t**2/2) / np.sqrt(2*np.pi)
func1_1 = lambda x: quad(integrand1_1, x**2, np.inf)[0] + np.exp(x) - 3*x**3 + 2*x + 4

# 그래프를 그려서 근이 있는 구간 찾기

# 정의역 설정
x1_1 = np.linspace(-8, 8, 100)

# 함수값 구하기
y1_1 = np.zeros_like(x1_1)
for i in range(len(x1_1)):
    y1_1[i] = func1_1(x1_1[i])

# 그래프 그리기
plt.plot(x1_1, y1_1)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of func1_1')
plt.grid()
plt.show()

# 그려진 그래프를 통해 F(x) = 0 인 부분은 두 구간 (-2,4)와 (6,8)에 있음을 확인
print("첫번째 근:", root_scalar(func1_1, method='bisect', bracket=(-2, 4)).root)
print("두번째 근:", root_scalar(func1_1, method='bisect', bracket=(6, 8)).root)
```

Q1-2. $$ F(x) = \int_{\sqrt{x}}^{x^2} \log(t) dt - \sqrt{x} - 3$$

```python
# your code here

# F(x) 식 세우기
def integrand1_2(t):
    return np.log(t)
func1_2 = lambda x: quad(integrand1_2, np.sqrt(x), x**2)[0] - np.sqrt(x) - 3

# 그래프를 그려서 근이 있는 구간 찾기

# 정의역 설정
x1_2 = np.linspace(1, 4, 100)

# 함수값 구하기
y1_2 = np.zeros_like(x1_2)
for i in range(len(x1_2)):
    y1_2[i] = func1_2(x1_2[i])

# 그래프 그리기
plt.plot(x1_2, y1_2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of func1_2')
plt.grid()
plt.show()

# 그려진 그래프를 통해 F(x) = 0 인 부분은 구간 (1,4)에 있음을 확인
print("근:", root_scalar(func1_2, method='brentq', bracket=(1, 4)).root)
```

Q1-3. 
$$
\begin{pmatrix}
F_1(x_1, x_2,\cdots,x_9) \\
\vdots \\
F_i(x_1, x_2,\cdots,x_9) \\
\vdots \\
F_9(x_1, x_2,\cdots,x_9)
\end{pmatrix}
=
\begin{pmatrix}
(3-2x_1)x_1 & -2x_2 & & & & & & +1 \\ 
& & & \vdots \\
& & -x_{i-1} & +(3-2x_i)x_i & -2x_{i+1} & & & +1 \\ 
& & & \vdots \\
& & & & & -x_8 & +(3-2x_9)x_9 & +1 \\ 
\end{pmatrix}
=
\begin{pmatrix}
0 \\
\vdots \\
0 \\
\vdots \\
0 \\
\end{pmatrix}
\text{for } i=2, \cdots\, 8
$$

```python
# your code here

# F(x) 식 세우기
def func1_3(x):
    F = np.array([])
    for i in range(0,9):
        if i == 0:
            F = np.append(F, (3-2*x[0])*x[0]-2*x[1]+1)
        elif i == 8:
            F = np.append(F, -x[7]+(3-2*x[8])*x[8]+1)
        else:
            F = np.append(F, -x[i-1]+(3-2*x[i])*x[i]-2*x[i+1]+1)
    return F

# 근 (x1,x2,...,x8,x9) 구하기
print('근 (x1,x2,...,x8,x9):', root(func1_3, [0 for i in range(0,9)], method='hybr').x)
```

## Consumer's Utiity Maximization Problem

There are many consumers in an economy who are different in labor productivity $x$. An individual consumer is endowed with one unit of time that can be used on either working to make money for consumption or enjoying as leisure. The wage per hour is given by $w$ that is determined in the market. All consumers receive the identical amount of lump-sum transfer $T$ from government. The utility function of the consumer is given by $ u(c, l) = c^{\frac{1}{3}} l^{\frac{2}{3}}$, where $c$ and $l$ denote consumption and leisure respectively. The consumer's problem can be written as follows:

$$
  \begin{align}
  \max_{c, l} c^{\frac{1}{3}} l^{\frac{2}{3}}\\
  \text{subject to}\quad c = wx(1-l) + T \\
  c \ge 0 \\
  0 \le l \le 1
  \end{align} 
$$
In the next two questions, we solve for the consumer's utility maximization problem in two different ways.


## Q2. Univariate function minimization


Q2-1. You may substitute the constraint (2) into the objective function (1) to make the problem an unconstrained univariable maximization problem with bounds (3) and (4). Write the Python function to compute the optimal consumption, leisure and labor supply $(c^*, l^*, N^s$) for an individual consumer with labor productivity $x$ given the lump-sum transfer $T$ and the market wage $w$. Within the function, solve for the utility maximization problem through the **Bounded** method.

```python
def bounded_method(x, w, T):
    # Bound method를 위한 함수 설정
    def objective_function(l):
        c = w * x * (1 - l) + T
        # Bound method는 최소값을 찾으므로 -1을 곱하여 최대값을 최소값으로 변경
        return -c**(1/3) * l**(2/3)

    # leisure의 범위(0,1)를 이용하여 최소값을 만드는 optimal leisure 찾기
    l_star = minimize_scalar(objective_function, method='bounded', bounds=(0,1)).x
    c_star = w * x * (1 - l_star) + T
    n_star = 1 - l_star
    
    # optimal consumption, leisure, labor supply 반환
    return c_star, l_star, n_star
```

Q2-2. For the time being, assume that $w=2$, $T=0.5$ and $\sigma_x=0.3$. Draw the graphs for the optimal consumption and leisure for 30 individuals whose log productivities are equally distributed over the range $\log{x} \in [-3\sigma_x, +3\sigma_x]$.

```python
# your code here

w = 2
sigma = 0.3

# Generate log productivity values for 30 individuals
log_x_values = np.linspace(-3*sigma, 3*sigma, 30)

# Convert log productivity values to productivity values
x_values = np.exp(log_x_values)

# Calculate optimal consumption and leisure for each individual
optimal_consumptions = []
optimal_leisures = []
for x in x_values:
    c_star, l_star, n_star = bounded_method(x, w, 0.5)
    optimal_consumptions.append(c_star)
    optimal_leisures.append(l_star)

# Plot optimal consumption and leisure for each individual
plt.plot(optimal_consumptions,optimal_leisures, 'o')
plt.xlabel('Consumption')
plt.ylabel('Leisure')
plt.title('Q2-2 Optimal Consumption and Leisure for 30 Individuals')
plt.show()
```

Q2-3. Draw the labor supply curve for the following two individuals as the market wage varies over the range of $w\in[1,4]$.

- Individual 1: $\log{x} = -\sigma_x$
- Individual 3: $\log{x} = \sigma_x$


```python
# your code here

# Parameters for the individuals
params = [-0.3, 0.3]

# Define the range of wages
w_range = np.linspace(1, 4, 100)

# Initialize an empty list to store the labor supply for each individual
labor_supply_list = []

# Loop over the range of wages and calculate the labor supply for each individual
for param in params:
    labor_supply_i_list = []
    for w in w_range:
        x = param
        labor_supply_i = bounded_method(np.exp(x), w, 0.5)[2]
        labor_supply_i_list.append(labor_supply_i)
    labor_supply_list.append(labor_supply_i_list)

# Plot the labor supply curve for each individual
plt.figure(figsize=(8, 6))
labels = ['Individual 1', 'Individual 3']
colors = ['blue', 'red']
for i, label in enumerate(labels):
    plt.plot(labor_supply_list[i], w_range, label=label, color=colors[i])
plt.title('Q2-3 Labor Supply Curve')
plt.xlabel('Labor Supply')
plt.ylabel('Wage')
plt.legend()
plt.show()
```

## Q3. Minimization of multivariate function with constraint

Q3-1. In this question, we will solve for the utility maximization as specified by equations (1)~(4). Write the Python function to compute the optimal consumption, leisure and labor supply $(c^*, l^*, N^s$) for an individual consumer with labor productivity $x$ given the lump-sum transfer $T$ and the market wage $w$. Within the function, solve for the utility maximization problem through the **Sequential Lest Squares Programming** algorithm.

```python
# your code here
def slsq(x, w, T):
    # slsq method를 위한 함수 설정
    def objective_function(a):
        # a[0] = c, a[1] = l
        # slsq method는 최소값을 찾으므로 -1을 곱하여 최대값을 최소값으로 변경
        return -a[0]**(1/3) * a[1]**(2/3)
    
    # Define the budget constraint: wx(1-l) + T - c
    eq_cons = {'type': 'eq', 'fun' : lambda a: np.array([w*x*(1-a[1]) + T - a[0]])}
    
    # Solve for the optimal consumption and leisure using the SLSQP algorithm
    result = minimize(objective_function, [0.5,1], method='SLSQP', bounds=((0, None), (0, 1)), constraints=[eq_cons])
    
    # Extract the optimal consumption, leisure, and labor supply
    c_star, l_star, n_star = result.x[0], result.x[1], 1-result.x[1]
    
    # Return the optimal values
    return c_star, l_star, n_star
```

Q3-3. Suppose that there are 10,000 individuals in the economy whose labor productivities are drawn from the following distributionis: $\log{x}\sim N(0, \sigma_x^2)$. Each individual behaves according to the utility maximization problem described above. Compute and draw histogram of the labor supplies of 10,000 individuals assuming the market wage being equal to 2.

```python
# your code here

# Define the parameters for the individuals
n = 10000
np.random.seed(2018121022)
x_dist = np.exp(np.random.normal(0, 0.3, n))

# Define the wage
w = 2

# Compute the labor supply for each individual
labor_supply = []
for i in range(n):
    x = x_dist[i]
    labor_supply_i = slsq(x, w, 0.5)[2]
    labor_supply.append(labor_supply_i)

# Draw histograms of the individual labor supplies of the two groups
plt.figure(figsize=(8, 6))
plt.hist(labor_supply, bins=50, alpha=0.5)
plt.title('Q3-3 Histogram of Optimal Labor Supply')
plt.xlabel('Labor Supply')
plt.ylabel('Frequency')
plt.show()
```

Q3-4. Draw the aggregate labor supply curve of the above economy as the market wage varies over the range of $w\in[1,4]$.

- create 15 grids of the market wages $w\in[1,4]$.
- for each $w$ compute labor supplies of 10,000 inividuals with different labor productivity as you did in Q3-3.
- the aggregate labor supply for a given $w$ is equal to the mean of labor supplies of the 10,000 individuals.
- plot the aggregate labor supply curve.

```python
# your code here

# Create 15 grids of the market wages
w_range = np.linspace(1, 4, 15)

# Calculate the aggregate labor supply for each wage in the range
labor_supply_agg = []
for w in w_range:
    labor_supply_w = []
    for i in range(n):
        x = x_dist[i]
        labor_supply_w.append(slsq(x, w, 0.5)[2])
    labor_supply_agg.append(np.mean(labor_supply_w))

# Plot the aggregate labor supply curve
plt.figure(figsize=(8, 6))
plt.plot(labor_supply_agg, w_range)
plt.title('Q3-4 Aggregate Labor Supply Curve')
plt.xlabel('Labor Supply')
plt.ylabel('Wage')
plt.show()
```
