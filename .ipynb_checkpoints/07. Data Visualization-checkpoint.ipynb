# Data Visualization

Data visualization is the graphical representation of data. 

It involves producing images to deliver important features of data to viewers. 

Visualizing data is an essential part of data analysis and machine learning. 

This lecture introduces Python library [Matplotlib](https://matplotlib.org).

- Line chart
- Bar chart
- Histogram
- Scatter plot
- Subplots
- 3D plots

````python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
````

## Line plot

The line plot is one of the simplest and most widely used data visualization techniques. 
A line plot displays information as a series of data points or markers connected by a straight line. 

We would like to show the population shares by age group of Korea in 1970~2020. Data are taken from KOSIS (https://kosis.kr/index/index.do).

````python
# population data

year = np.array([1970, 1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015, 2020])
share_0_14 = np.array([42.5, 38.6, 34.0, 30.2, 25.6, 23.4, 21.1, 19.1, 16.1, 13.8, 12.2])
share_15_64 = np.array([54.4, 58.0, 62.2, 65.6, 69.3, 70.7, 71.7, 71.9, 73.1, 73.4, 72.1])
share_65_99 = np.array([3.1, 3.5, 3.8, 4.3, 5.1, 5.9, 7.2, 9.0, 10.8, 12.8, 15.7])
````

### Quick plotting

````python
# plot a single line in a chart

plt.plot(year, share_0_14)
````

````python
# plot multiple lines in a chart

plt.plot(year, share_0_14)
plt.plot(year, share_15_64)
plt.plot(year, share_65_99)
````

````python
# plot with more information

plt.figure(figsize=(9,6))
plt.plot(year, share_0_14)
plt.plot(year, share_15_64)
plt.plot(year, share_65_99)

plt.xlabel('Year')
plt.ylabel('Share (%)')
plt.title('Composition of Population by Age Group')
plt.legend(['Young: 0-14', 'Working: 15-64', 'Old: 65+'])
plt.show()
````

### Decorating line plot

There are many other options in the Matplotlib libary to make the chart more visible. Examples include:

- figure size, face color
- line style 
- line color
- line width
- line markers
  * marker
  * marker size
  * marker edge color
  * marker edge width
  * marker face color
- legend
  * location
  * font size

For more information on the options for plot refer to https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

````python
# chart decorations

plt.figure(figsize=(9,6), facecolor='white')
plt.plot(year, share_0_14, color='red', linestyle='-', linewidth=2, marker='o', markersize='10', markerfacecolor='red', markeredgecolor='black', markeredgewidth='2', alpha=1, label='Young: 0-14')
plt.plot(year, share_15_64, c='g', ls='--', lw=4, marker='v', ms='10', mec='k', mew='2', alpha=0.6, label='Working: 15-64')
plt.plot(year, share_65_99, 'b-.^', alpha=0.3, label='Old: 65 over')

plt.xlabel('Year', fontsize=15)
plt.ylabel('Share (%)', fontsize=15)
plt.title('Composition of Population by Age Group', fontsize=20)
plt.legend(loc='center right', fontsize=15)
plt.grid(True)
plt.tight_layout()
plt.savefig('Data_Fig/Line_Plot.png')
plt.show()
````

## Bar chart

Another type of data visualization technique widely used is the bar chart, which shows the data using bars with different heights. Below we will show the changes in the composition of population in Korea using a bar chart.

````python
# single series in a bar chart

plt.figure(figsize=(9,6))
plt.bar(year, share_0_14, alpha=1, label='Young: 0-14')
plt.show()
````

````python
# multiple series in a bar chart

plt.figure(figsize=(9,6), facecolor='white')

plt.bar(year, share_0_14, color='red', label='Young: 0-14')
plt.bar(year, share_15_64, color='g', label='Working: 15-64')
plt.bar(year, share_65_99, color='b', label='Old: 65 over')
plt.legend()
plt.show()
````

````python
# multiple series in a bar chart: a little trick

year_width = 1.0
width = 1.0

plt.figure(figsize=(9,6))
plt.bar(year-year_width, share_0_14, color='red', width=width, label='Young: 0-14')
plt.bar(year, share_15_64, color='green', width=0.8*width, label='Working: 15-64')
plt.bar(year+year_width, share_65_99, color='blue', width=0.8*width, label='Old: 65 over')

plt.xlabel('Year', fontsize=15)
plt.ylabel('Share (%)', fontsize=15)
plt.title('Composition of Population by Age Group', fontsize=20)
plt.legend()
plt.tight_layout()
plt.show()
````

````python
# multiple series in a horizontal bar chart

year_height = 1.
height = 1.

plt.figure(figsize=(9,9))
plt.barh(year-year_height, share_0_14, color='red', height=0.8*height, label='Young: 0-14')
plt.barh(year, share_15_64, color='green', height=0.8*height, label='Working: 15-64')
plt.barh(year+year_height, share_65_99, color='blue', height=0.8*height, label='Old: 65 over')

plt.xlabel('Year', fontsize=15)
plt.ylabel('Share (%)', fontsize=15)
plt.title('Composition of Population by Age Group', fontsize=20)
plt.legend()
plt.tight_layout()
plt.show()
````

````python
# stack bars on each year

width=3

plt.figure(figsize=(9,6))
plt.bar(year, share_0_14, color='red', width=width, label='Young: 0-14')
plt.bar(year, share_15_64, bottom=share_0_14, width=width, color='g', label='Working: 15-64')
plt.bar(year, share_65_99, bottom=share_0_14+share_15_64, width=width, color='b', label='Old: 65 over')

plt.xlabel('Year', fontsize=15)
plt.ylabel('Share (%)', fontsize=15)
plt.title('Composition of Population by Age Group', fontsize=20)
plt.legend(loc='center right')
plt.tight_layout()
plt.show()
````

## Stack Plot

````python
# use stackplot

plt.figure(figsize=(9,6))
plt.stackplot(year, share_0_14, share_15_64, share_65_99, colors = ['red', 'green', 'blue'], labels=['Young: 0-14', 'Working: 15-64', 'Old: 65 over'])

plt.xlabel('Year', fontsize=15)
plt.ylabel('Share (%)', fontsize=15)
plt.title('Composition of Population by Age Group', fontsize=20)
plt.legend(loc='center right')
plt.tight_layout()
plt.show()
````

## Histogram

A histogram is often used to display the frequencies of data points of a variable. Below we will create a hisogram of 10,000 random numbers drawn from a standard normal distribution, $X\sim N(0,1)$.

### Quick plotting

````python
# generate randon numbes

X = np.random.randn(10000)
````

````python
# simple histogram

plt.figure(figsize=(9,6))
plt.hist(X, bins=50)

plt.title(r'Standard Normal Distribution: $X\sim N(0,1)$', fontsize=20)
plt.xlabel('X', fontsize=15)
plt.ylabel('No. of obs', fontsize=15)
plt.tight_layout()
plt.show()
````

### Decorating histogram

Matplotlib provides with many options to improve the clarity and visibility of histograms. Some of them are introduced below. For more information on other options for histogram, refer to https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html

````python
# add color, edge color of bars, density (pdf)

plt.figure(figsize=(9,6))
plt.hist(X, bins=50, color='green', edgecolor='black', width=0.1, density=True)
plt.axvline(X.mean(), color='red', label='Mean', linewidth=2)

plt.title('Standard Normal Distribution: $X\sim N(0,1)$', fontsize=20)
plt.xlabel('X', fontsize=15)
plt.ylabel('PDF(x)', fontsize=15)
plt.legend(fontsize=15)
plt.tight_layout()
plt.show()
````

````python
# make cumulative distribution function

plt.figure(figsize=(9,6))
plt.hist(X, bins=50, color='yellow', edgecolor='red', density=True, cumulative=True)

plt.title('Standard Normal Distribution: $X\sim N(0,1)$', fontsize=20)
plt.xlabel('X', fontsize=15)
plt.ylabel('CDF(x)', fontsize=15)
plt.tight_layout()
plt.show()
````

````python
# change the orientation of the histogram

plt.figure(figsize=(6,9))
plt.hist(X, bins=50, color='magenta', edgecolor='blue', density=False, cumulative=False, orientation='horizontal')

plt.title('Standard Normal Distribution: $X\sim N(0,1)$', fontsize=20)
plt.ylabel('X', fontsize=15)
plt.xlabel('No. of obs', fontsize=15)
plt.tight_layout()
plt.show()
````

## Scatter Plot

A scatter plots data points of two variables on a two dimensional grid. In addition, more variables can be employed to determine the size or color of the points. Below is an example of scatter plot.

````python
# generate a set of random data set

nobs = 100
scat_X = np.random.randn(nobs)
scat_Y = np.random.randn(nobs)
scat_sizes = np.random.rand(nobs)*1000
scat_colors = np.random.rand(nobs)*100
````

````python
# scatter plot

plt.figure(figsize=(9,6))
plt.scatter(scat_X, scat_Y, s=scat_sizes, c=scat_colors, cmap='Blues', marker='o', edgecolor='black', linewidth=1, alpha=0.9)
cbar=plt.colorbar()
cbar.set_label('Random Size', fontsize=12)

plt.title('Scatter Plot of Random Coordinates', fontsize=20)
plt.xlabel('X', fontsize=15)
plt.ylabel('Y', fontsize=15)
plt.tight_layout()
plt.show()
````

## Plotting multiple charts in a figure

Matplotlib also support plotting multiple charts in one figure, using `plt.subplots`, which returns a set of axes for plotting. The basic syntax is as follows:

```
fig, axes = plt.subplots(n_rows, n_cols)

axes[0,0].plots(x,y)
axes[0,1].bar(x,y)
:
:
```

The following is the simplified illustration of the structure of a Matplotlib figure and axes: 

<img src="Data_Fig/matplotlib_figure_axes_axis.png" style="width:50%;">

Below, we will put together in one figure the various types of charts we have covered above each of which is put in a axes . For more information on `subplots` refer to https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html

````python
# set figure frame and axes

fig, axes = plt.subplots(3, 3, figsize=(15, 12))
````

````python
# set figure and axes

fig, axes = plt.subplots(3, 3, figsize=(15, 12))

# add charts on each axes

# line plot

axes[0,0].plot(year, share_0_14, color='red', linestyle='-', linewidth=2, marker='o', markersize='10', markerfacecolor='red', markeredgecolor='black', markeredgewidth='2', alpha=1, label='Young: 0-14')
axes[0,0].plot(year, share_15_64, c='g', ls='--', lw=4, marker='v', ms='10', mec='k', mew='2', alpha=0.6, label='Working: 15-64')
axes[0,0].plot(year, share_65_99, 'b-.^', alpha=0.3, label='Old: 65 over')
axes[0,0].set_xlabel('Year')
axes[0,0].set_ylabel('Share (%)')
axes[0,0].set_title('Line Plot')
axes[0,0].legend(loc='center right')
axes[0,0].grid(True)

# bar chart

width = 1.3
axes[0,1].bar(year-width, share_0_14, color='red', width=0.8*width, label='Young: 0-14')
axes[0,1].bar(year, share_15_64, color='green', width=0.8*width, label='Working: 15-64')
axes[0,1].bar(year+width, share_65_99, color='blue', width=0.8*width, label='Old: 65 over')
axes[0,1].set_xlabel('Year', fontsize=10)
axes[0,1].set_ylabel('Share (%)', fontsize=10)
axes[0,1].set_title('Bar Chart', fontsize=12)

# horizontal bar chart

height = 1.3
axes[0,2].barh(year-height, share_0_14, color='red', height=0.8*height, label='Young: 0-14')
axes[0,2].barh(year, share_15_64, color='green', height=0.8*height, label='Working: 15-64')
axes[0,2].barh(year+height, share_65_99, color='blue', height=0.8*height, label='Old: 65 over')
axes[0,2].set_xlabel('Year')
axes[0,2].set_ylabel('Share (%)')
axes[0,2].set_title('Horizontal Bar Chart')

# stacked bar chart

width=3
axes[1,0].bar(year, share_0_14, color='red', width=width, label='Young: 0-14')
axes[1,0].bar(year, share_15_64, bottom=share_0_14, width=width, color='g', label='Working: 15-64')
axes[1,0].bar(year, share_65_99, bottom=share_0_14+share_15_64, width=width, color='b', label='Old: 65 over')
axes[1,0].set_xlabel('Year')
axes[1,0].set_ylabel('Share (%)')
axes[1,0].set_title('Stacked Bar Chart')

# stackplot

axes[1,1].stackplot(year, share_0_14, share_15_64, share_65_99, colors = ['red', 'green', 'blue'], labels=['Young: 0-14', 'Working: 15-64', 'Old: 65 over'])
axes[1,1].set_xlabel('Year')
axes[1,1].set_ylabel('Share (%)')
axes[1,1].set_title('Stackplot')

# histogram

axes[1,2].hist(X, bins=50, color='green', edgecolor='black', width=0.1, density=True)
axes[1,2].axvline(X.mean(), color='red', label='Mean', linewidth=2)
axes[1,2].set_title('Histogram')
axes[1,2].set_xlabel('X')
axes[1,2].set_ylabel('PDF(x)')

# cumulative histogram

axes[2,0].hist(X, bins=50, color='yellow', edgecolor='red', density=True, cumulative=True)
axes[2,0].set_title('Cumulative Histogram')
axes[2,0].set_xlabel('X', fontsize=15)
axes[2,0].set_ylabel('CDF(x)')

# horizontal histogram

axes[2,1].hist(X, bins=50, color='magenta', edgecolor='blue', density=False, cumulative=False, orientation='horizontal')
axes[2,1].set_title('Horizontal Histogram')
axes[2,1].set_xlabel('X')
axes[2,1].set_ylabel('No. of obs')

# scatter plot

axes[2,2].scatter(scat_X, scat_Y, s=scat_sizes, c=scat_colors, cmap='Blues', marker='o', edgecolor='black', linewidth=1, alpha=0.9)
axes[2,2].set_title('Scatter Plot of Random Coordinates')
axes[2,2].set_xlabel('X')
axes[2,2].set_ylabel('Y')

plt.tight_layout()
plt.show()
````

## 3D Surface plot

Suppose that the consumer derives utility from consumption and leisure and the utility function is given as follows: 
$$
U(c, l) = \frac{c^{1-\sigma}-1}{1-\sigma} + \frac{l^{1-\gamma}-1}{1-\gamma}
$$

The utility can be displayed on 3-D surface plot as below.

````python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

def U(c,l, sigma=2, gamma=1.5):
    return (c**(1-sigma) - 1)/(1-sigma) + (l**(1-gamma) - 1)/(1-gamma)

cgrid = np.linspace(0.1, 1, 30)
lgrid = np.linspace(0.1, 1, 30)
cons, leisure = np.meshgrid(cgrid, lgrid)
utility = U(cons, leisure)

fig = plt.figure(figsize=(18,6))

# 3D surface

ax = fig.add_subplot(121, projection='3d')
surface = ax.plot_surface(cons, leisure, utility, rstride=1, cstride=1, cmap=cm.jet)
ax.set_xlabel('consumption')
ax.set_ylabel('leisure')
ax.set_title('Utility Surface (3D)')
plt.colorbar(surface)

# 2D contour

ax = fig.add_subplot(122)
contour = ax.contourf(cons, leisure, utility, cmap=cm.jet)
ax.set_xlabel('consumption')
ax.set_ylabel('leisure')
ax.set_title('Indifference Curve (2D)')
plt.colorbar(contour)
plt.show()
````
